## Function

Convert plain text to ascii values ---> Convert ascii values to 8-bit binary number ---> combine all the binary number and make it into block of 64 bits.

<div>

</div>

```python
def expansion_permutation_box_operation(text):
    
    text = " " + text
    s = ""
    s = s + text[32]+text[1:5]+text[5]
    s = s + text[4]+text[5:9]+text[9]
    s = s + text[8]+text[9:13]+text[13]
    s = s + text[12]+text[13:17]+text[17]
    s = s + text[16]+text[17:21]+text[21]
    s = s + text[20]+text[21:25]+text[25]
    s = s + text[24]+text[25:29]+text[29]
    s = s + text[28]+text[29:33]+text[1]
    
    print("\nOutput of Expansion Permutation Box Operation =",s)
    return s

def process(plain_text):
    
    bpt = ""
    for i in plain_text:
        x = bin(ord(i)).replace("0b","")
        if(len(x)<8):
            x = "0"*(8-len(x)) + x
        bpt = bpt + x
            
    print("\n\nBlocks of 64-bits : \n")
    output = []
    for i in range(0,len(bpt),64):
        print(bpt[i:i+64])
        output.append(bpt[i:i+64]+"\n")
    
    output.append("Output of Expansion Permutation Box Operation = "+expansion_permutation_box_operation(bpt[32:64]))
    return output
```
