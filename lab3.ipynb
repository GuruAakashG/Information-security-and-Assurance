### Getting details and check intrusion of pcap file

```python
def check(n,pkt):
    src = []
    dst = []
    proto = []
    ty = []
    for i in range(0,n):
        try:
            if(pkt[i]['Ethernet'].type==2048):
                src.append(pkt[i]['IP'].src)
                dst.append(pkt[i]['IP'].dst)
                ty.append("IPv4")
                if(pkt[i]['IP'].proto==1):
                    proto.append("ICMP")
                elif(pkt[i]['IP'].proto==6):
                    proto.append("TCP")
                elif(pkt[i]['IP'].proto==17):
                    proto.append("UDP")
                else:
                    proto.append("Other")
            else:
                src.append("NA")
                dst.append("NA")
                ty.append(pkt[i]['Ethernet'].type)
                proto.append("NA")
        except:
            src.append("NA")
            dst.append("NA")
            ty.append("Other")
            proto.append("NA")
    
    s = []
    count_intrusion = 0
    for i in range(0,n):
        if(src[i]==dst[i]):
            count_intrusion = count_intrusion + 1
            print("Analyzed Packet is Intrusion Packet")
            s.append("Analyzed Packet is Intrusion Packet\n")
        elif(src[i]=="255.255.255.255" or dst[i]=="255.255.255.255"):
            count_intrusion = count_intrusion + 1
            print("Analyzed Packet is Intrusion Packet")
            s.append("Analyzed Packet is Intrusion Packet\n")
        elif(proto[i]=="ICMP"):
            count_intrusion = count_intrusion + 1
            print("Analyzed Packet is Intrusion Packet")
            s.append("Analyzed Packet is Intrusion Packet\n")
        else:
            print("Analyzed Packet is Not-Intrusion Packet")
            s.append("Analyzed Packet is Not-Intrusion Packet\n")
        print(pkt[i])
        s.append(str(pkt[i])+"\n")
            
    print("\nTotal number of packets checked =",n)
    print("Total number of intrusion packets detected =",count_intrusion)
    s.append("\nTotal number of packets checked = "+str(n)+"\n")
    s.append("Total number of intrusion packets detected = "+str(count_intrusion)+"\n")
    
    return s

def check_intrusion(n,pcap_filename):
    pkt = rdpcap(pcap_filename)
    if(len(pkt)<n):
        s = check(len(pkt),pkt)
        print("Only",len(pkt),"packet is available in pcap file")
    else:
        s = check(n,pkt)
    return s
```

### Sample Testcase

```python
def store_output_file(filename,output):
    file = open(filename,"w")
    file.writelines(output)
    file.close()
```

```python
n = 35
s = check_intrusion(n,"Sample-Testcase.pcap")
store_output_file("Sample-Testcase.txt",s)
```
